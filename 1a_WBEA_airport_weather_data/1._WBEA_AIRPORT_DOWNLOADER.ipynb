```python
#Script to download data from all cities in WBEA_AIRPORTS.txt
#Run in order by cell top to bottom

#Clears all previous information
import readline 
readline.clear_history()

#Importing needed libraries 
import pandas as pd #Import pandas package for dataframes
import os #Import for setting working dirrectory
import requests #Import for downloading data from database
```

```python

os.chdir('/Users/milsbeary/Desktop/Miles NSERC Alliance Missions /Air Quality Monotoring of Polutents/Land Station/WBEA_Airport_Weather_Data') #Set working dirrectory 

cities = pd.DataFrame(pd.read_csv('WBEA_AIRPORTS.csv', delimiter = None)).dropna(axis=1, how='all') #Import WBEA_AIRPORTS as 
#print(cities) #View city list


#Adding back zeros that excell deleted infront of the USAF and WBAN numbers
for i in range(len(cities)): #Note that the df starts at element zero 
    
    #Extracting parameters from dataframe
    usaf_unmodified = cities.loc[i , "USAF"] #Extracts the value by row number and collumn name
    #print(usaf_unmodified) #View extracted parameter
    wban_unmodified = cities.loc[i , "WBAN"]
    #print(wban_unmodified)
    
    #Adding a zero infront of short usaf number
    if usaf_unmodified < 100000:
        cities.loc[i , "USAF"] = str(usaf_unmodified).zfill(6) #Appends a zero at the 6th element infront of the usaf number
        #print(str(usaf_unmodified).zfill(5))
    
    #Adding a zero infronto of short wban number 
    if wban_unmodified < 10000: 
        cities.loc[i , "WBAN"] = str(wban_unmodified).zfill(5) #Appends a zero at the 5th element infront of the wban number
        #print(str(wban_unmodified).zfill(5))


print(cities) #View entire dataframe to make sure everyting is correct
    

```

```python
#Extracting and saving csvs from the NOAA website

# Base URL and filename
url_base_hourly = "https://www.ncei.noaa.gov/data/global-hourly/access/" #Hourly data path (icludes 3 hour data)
url_base = "https://www.ncei.noaa.gov/data/global-summary-of-the-day/access/" #Daily data path 

for i in range(0, len(cities)): #Itterates over each city

    print(i) #Gives itteration the loop is on
    
    #Extracting needed parameters from dataframe
    usaf = cities.loc[i , "USAF"]
    #print(usaf) 
    wban = cities.loc[i , "WBAN"]
    #print(wban)
    startyear = str(cities.loc[i , "BEGIN"])[:4] #Extracts the year from the date string by taking the first four elements
    #print(startyear)
    endyear = str(cities.loc[i , "END"])[:4]  
    #print(endyear)
    
    #Defines and locates filename
    filename_noaa = str(usaf) + str(wban) + ".csv" #Define CSV file name (Should match what is on the server)
    #print(filename_noaa)
    
    #Downloading files to computer
    for yr in range(int(startyear), int(endyear)+1): #Sets the year values as integers becuase they are initally strings
        #print(yr) #View index 
        
        #Set exact path to match where eah file is on the NOAA website 
        url_loop_daily = url_base + str(yr) + "/" + filename_noaa
        #print(url_loop_daily)
        url_loop_hourly = url_base_hourly + str(yr) + "/" + filename_noaa 
        #print(url_loop_hourly)
        
        #Identifies file on website
        csv_current_daily = requests.get(url_loop_daily, allow_redirects = True)
        csv_current_hourly = requests.get(url_loop_hourly, allow_redirects = True)
        
        #Sets the file name of the local imported file
        filename_local_daily =  str(usaf) + str(wban) + "-" + str(yr) + "DAILY.csv" 
        filename_local_hourly = str(usaf) + str(wban) + "-" + str(yr) + "HOURLY.csv"
        
       
        os.chdir('/Users/milsbeary/Desktop/Miles NSERC Alliance Missions /Air Quality Monotoring of Polutents/Land Station/WBEA_Airport_Weather_Data/Daily') #Sets path for daily data csv deposit  
        open(filename_local_daily, "wb").write(csv_current_daily.content) #Write a csv for daily data
        
        os.chdir('/Users/milsbeary/Desktop/Miles NSERC Alliance Missions /Air Quality Monotoring of Polutents/Land Station/WBEA_Airport_Weather_Data/Hourly') #Sets path for hourly data csv deposit
        open(filename_local_hourly, "wb").write(csv_current_hourly.content) #w opens the file for writing and b puts it in bianary mode
    
    #end of nested loop
#end of origional loop
#Set up a count here 
```

```python
#Deleting all empty files in the hourly folder

os.chdir('/Users/milsbeary/Desktop/Miles NSERC Alliance Missions /Air Quality Monotoring of Polutents/Land Station/WBEA_Airport_Weather_Data/Hourly') #Setting our dirrectory
hour_dir = os.listdir('/Users/milsbeary/Desktop/Miles NSERC Alliance Missions /Air Quality Monotoring of Polutents/Land Station/WBEA_Airport_Weather_Data/Hourly') #Defining the daily dirrectory 
h_dir_len = len(hour_dir) #Setting a variable for the amount of files in the dirrectory

for i in range(h_dir_len):
    
    print(i) 

    h_size = os.stat(hour_dir[i]).st_size #Extract file size
    #print(h_size) #Making sure it works
    
    if h_size == 244: #Size of empty file 
    
        os.remove(hour_dir[i]) #remove the empty file
    
    #End of if else statement 

#End of for loop  
```
